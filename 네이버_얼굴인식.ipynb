```python
import os               # 사진 파일을 읽어오기 위한 모듈
import sys              # 사진 파일을 읽어오기 위한 모듈
import requests         # 네이버에 사진 보내고 받기

client_id = 'hupTgF5IoLZIOTc5ktI0'      # 네이버 개발자 센터에서 발급
client_secret = 'gAP6jTA3UI'


url = "https://openapi.naver.com/v1/vision/face" # 얼굴감지
# url = "https://openapi.naver.com/v1/vision/celebrity" # 유명인 얼굴인식
files = {'image': open('photo.jpg', 'rb')}
headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret }
response = requests.post(url,  files=files, headers=headers)

print(response.text)
```

```python
import json
parsed = json.loads(response.text)    # 자바스크립트의 json 을 파이썬의 딕셔너리로 변환
print(parsed)
```

```python
import matplotlib.patches as patches          # 사진 수정 패치

img = mpimg.imread('photo.jpg')               # 수정할 사진 읽어오기  
fig, ax = plt.subplots(figsize=(10,10))       # 수정할 사진 확대
ax.imshow(img)                                # 사진 보기

x,y,w,h = person['roi'].values()              # 얼굴 정보 roi  얼굴의 위치와 크기

re_face = patches.Rectangle((x,y), w, h, edgecolor='r', facecolor='none')   # 사진 수정

gender, gender_c = person['gender'].values()
age, age_c = person['age'].values()

anno = gender + ',' + (str)(gender_c) + '\n' + age + ',' + (str)(age_c)
plt.text(x,y+h+70, anno, color='blue')
ax.add_patch(re_face)                         # 수정된 사진 보기
```
