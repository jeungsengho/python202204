```python
# 딕셔너리로 그래프 그리기

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}
plt.plot('data_x','data_y', data=data_dict)   # 딕셔너리로 데이터 삽입
plt.show()
```

```python
data_dict = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}

plt.plot('data_x','data_y', data=data_dict)
plt.xlabel('X축')
plt.ylabel('Y축')

plt.show()
```

```python
# 축 설정
data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}

plt.plot('data_x','data_y', data=data_dict)

# 축의 이름 설정
plt.xlabel('X축', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r', 'size':14})

plt.show()
```

```python
# 범례 설정 legend

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}

plt.plot('data_x','data_y', data=data_dict, label = 'Price ($)')
plt.xlabel('X축', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r', 'size':14})

plt.legend()  # 범례 표시
plt.show()

# 그래프 영역에 범례를 나타내려면 plot 함수에 label 문자열을 저장하고
# legend() 함수를 사용하여 표시할 수 있다.
```

```python
bitcoin = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
ethereum = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=bitcoin, label = 'BTC')
plt.plot('data_x','data_y', data=ethereum, label = 'ETH')

plt.xlabel('월', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('가격', fontdict={'color':'r', 'size':14})

plt.legend(loc = 0)  
# 범례의 위치는 지정하지 않으면 적당한 위치로 자동 설정된다.
# 10 - 정가운데 9 - 위에서 가운데 5 - 우측 가운데 3 - 좌측 하단 1 - 우측 상단
plt.show()
```

```python
bitcoin = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
ethereum = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=bitcoin, label = 'BTC')
plt.plot('data_x','data_y', data=ethereum, label = 'ETH')

plt.xlabel('월', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('가격', fontdict={'color':'r', 'size':14})

plt.legend(loc = 0, ncol=2)   # 범례를 나타낼 때 열의 개수를 지정 ncol

plt.show()
```

```python
bitcoin = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
ethereum = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=bitcoin, label = 'BTC')
plt.plot('data_x','data_y', data=ethereum, label = 'ETH')

plt.xlabel('월', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('가격', fontdict={'color':'r', 'size':14})

plt.legend(loc = 0, ncol=2, fontsize=10, shadow=True)
# fontsize=10으로 범례의 글자 크기 지정, shadow=True 범례의 테두리 그림자 설정
plt.show()
```

```python
bitcoin = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
ethereum = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=bitcoin, label = 'BTC')
plt.plot('data_x','data_y', data=ethereum, label = 'ETH')

plt.xlabel('월', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('가격', fontdict={'color':'r', 'size':14})

plt.legend(loc = 0, ncol=2, fontsize=10, frameon=False)
# frameon-False 범례의 테두리 삭제
plt.show()
```

```python
# 여러 그래프 한번에 그리기

# 먼저 한개의 그래프 완성
plt.plot([1,2,3,4],[2,2.5,3.3,4.3])

plt.show()
```

```python
plt.plot([1,2,3,4],[2,2.5,3.3,4.3], 'r*-', label = 'BTC')  # 빨강
plt.plot([1,2,3,4],[2,3,3,6], 'go-', label = 'ETH')        # 녹색
plt.plot([1,2,3,4],[2,3,5,10], 'b^-', label = 'ADA')       # 파랑

plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[2,2.5,3.3,4.3], 'r*-', label = 'BTC')  # 빨강
plt.plot([1,2,3,4],[2,3,3,6], 'go-', label = 'ETH')        # 녹색
plt.plot([1,2,3,4],[2,3,5,10], 'b^-', label = 'ADA')       # 파랑

plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[2,2.5,3.3,4.3], 'r*-', markersize=20, label = 'BTC')  # 빨강
plt.plot([1,2,3,4],[2,3,3,6], 'go-', label = 'ETH')        # 녹색
plt.plot([1,2,3,4],[2,3,5,10], 'b^-', linewidth=10, label = 'ADA')       # 파랑

plt.grid(True)
plt.legend()
plt.show()
```

```python
# 그래프의 타이틀 제목 설정

plt.plot([1,2,3,4],[2,2.5,3.3,4.3], 'r*-', markersize=20, label = 'BTC')  # 빨강
plt.plot([1,2,3,4],[2,3,3,6], 'go-', label = 'ETH')        # 녹색
plt.plot([1,2,3,4],[2,3,5,10], 'b^-', linewidth=10, label = 'ADA')       # 파랑

plt.title('코인 가격 증가율')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 막대 그래프 - 값을 비교하는 용도로 주로 사용된다.

years = ['2019', '2020', '2021']
values = [100, 400, 900]

plt.bar(years, values)

plt.show()
```

```python
years = ['2019', '2020', '2021']
values = [100, 400, 900]

plt.bar(years, values, width=0.6)

plt.show()
```

```python
years = ['2019', '2020', '2021']
values = [100, 400, 900]

plt.bar(years, values, width=0.6, color='green')

plt.show()
```

```python
# 가로 막대 그래프

years = ['2019', '2020', '2021']
values = [800, 700, 900]

plt.barh(years, values)

plt.show()
```

```python
years = ['2019', '2020', '2021']
values = [800, 700, 900]

plt.barh(years, values, color='green', height=0.5, edgecolor='black')

plt.show()
```

```python
months = ['1월','2월','3월','4월','5월']
price = [8000, 7000, 6000, 7000, 4000]

plt.bar(months, price, color='red', width=0.3, edgecolor='black')

plt.title('비트코인 가격 변동 추이')
plt.xlabel('월')
plt.ylabel('가격')
plt.show()
```
