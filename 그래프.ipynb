```python
# 범례 설정 legend

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}

plt.plot('data_x','data_y', data=data_dict, label = 'Price ($)')
plt.xlabel('X축', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r', 'size':14})

plt.legend()  # 범례 표시
plt.show()

# 그래프 영역에 범례를 나타내려면 plot 함수에 label 문자열을 저장하고
# legend() 함수를 사용하여 표시할 수 있다.
```

```python
bitcoin = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
ethereum = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=bitcoin, label = 'BTC')
plt.plot('data_x','data_y', data=ethereum, label = 'ETH')

plt.xlabel('월', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('가격', fontdict={'color':'r', 'size':14})

plt.legend(loc = 0)  
# 범례의 위치는 지정하지 않으면 적당한 위치로 자동 설정된다.
# 10 - 정가운데 9 - 위에서 가운데 5 - 우측 가운데 3 - 좌측 하단 1 - 우측 상단
plt.show()
```

```python
bitcoin = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
ethereum = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=bitcoin, label = 'BTC')
plt.plot('data_x','data_y', data=ethereum, label = 'ETH')

plt.xlabel('월', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('가격', fontdict={'color':'r', 'size':14})

plt.legend(loc = 0, ncol=2)   # 범례를 나타낼 때 열의 개수를 지정 ncol

plt.show()
```

```python
bitcoin = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
ethereum = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=bitcoin, label = 'BTC')
plt.plot('data_x','data_y', data=ethereum, label = 'ETH')

plt.xlabel('월', fontdict={'color':'b', 'size':'xx-large'})
plt.ylabel('가격', fontdict={'color':'r', 'size':14})

plt.legend(loc = 0, ncol=2, fontsize=10, shadow=True)
# fontsize=10으로 범례의 글자 크기 지정, shadow=True 범례의 테두리 그림자 설정
plt.show()
```

```python
# 산점도 scatter plot 분포 등을 표현할 때 사용

import numpy as np

np.random.seed(1)   # 난수를 발생시킬 때 일정한 패턴으로만 발생하도록 고정
                    # 예를 들어 1~10 사이의 난수를 3개 발생시킬때 항상 4,6,8로만
                    #           난수가 생성되도록 고정 시킨다.
                  
n = 50

x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x,y)
plt.show()
```

```python
np.random.seed(1) # 난수 발생 패턴을 일정하게 고정시킨다.

n = 50

x = np.random.rand(n)
y = np.random.rand(n)
size = (30 * np.random.rand(n)) ** 2    # 크기를 설정할 난수 발생

plt.scatter(x,y, s=size)
plt.show()
```

```python
np.random.seed(1) # 난수 발생 패턴을 일정하게 유지

n = 50

x = np.random.rand(n)                     # 랜덤하게 생성된 x축 값
y = np.random.rand(n)                     # 랜덤하게 생성된 y축 값 
size = (30 * np.random.rand(n)) ** 2      # 랜덤하게 생성된 값의 크기 (크기의 차이를 크게 만들기 위해 30을 곱하고 제곱)
color = np.random.rand(n)                 # 랜덤하게 생성된 값의 색깔

plt.scatter(x,y, s=size, c=color, alpha=0.4)  # alpha 투명도
plt.show()
```

```python
np.random.seed(1) # 난수 발생 패턴을 일정하게 유지

n = 50

x = np.random.rand(n)                     # 랜덤하게 생성된 x축 값
y = np.random.rand(n)                     # 랜덤하게 생성된 y축 값 
size = (30 * np.random.rand(n)) ** 2      # 랜덤하게 생성된 값의 크기 (크기의 차이를 크게 만들기 위해 30을 곱하고 제곱)
color = np.random.rand(n)                 # 랜덤하게 생성된 값의 색깔

plt.scatter(x,y, s=size, c=color, alpha=0.4, cmap='Set1')  # cmap 색상표
plt.colorbar()    # 색상의 범위를 시각적으로 표현
plt.show()
```

```python
# Wordcloud

# !pip install Wordcloud

import wordcloud
import matplotlib.pyplot as plt

words = {
    'Python':10,
    'Java':5,
    'C':7,
    'C++':9,
    'JSP':12
}
wc = wordcloud.WordCloud()
cloud = wc.generate_from_frequencies(words)
plt.figure()
plt.imshow(cloud)
plt.show()
```
